## Notebook summary: 

This notebook covers some of the preprocessing steps for image processing: 
  - Converting image to pixel array 
  - Converting RGB image to grayscale image
  - Normalising the pixel array

This is done using Matplotlib, Numpy and Pillow. Note that in reality, only one of these is required and this is only done as a comparison for these methods.

```python
# import pillow for image manipulation
import PIL
from PIL import Image

# Check Pillow version number
print('Pillow Version:', PIL.__version__)

# import relevant Matplotlib modules 
from matplotlib import image
from matplotlib import pyplot 

# import image-to-array from numpy library
import numpy as np
from numpy import asarray
```

## Using Matplotlib library to generate pixel array from JPEG file:

```python
# load image as pixel array
data = image.imread('horsmil.jpg')

# summarize shape of the pixel array
print('Data type of pixel array is: ', data.dtype)
print('Shape of dataset is: ',data.shape)

# display the array of pixels as an image
pyplot.imshow(data)
pyplot.show()

data
```

## Using Pillow library to generate pixel array from JPEG file:

```python
# Load image using Pillow 
img = Image.open('horsmil.jpg')

# Image summary 
print('Image format is: ', img.format)
print('Image mode is: ',img.mode)
print('Image mode is: ',img.size)

# Open image 
img.show()

# Generate pixel array from image 
pixels = list(img.getdata())

# Check to see if array is in correct format
print('\nLength of pixel array using pillow library: ', len(pixels))
```

### Convert RGB image to Grayscale using Pillow:

```python
# Convert to grayscale using pillow 
img_gray = Image.open('horsmil.jpg').convert('L')
img_gray.save('grayscale_horsmil.jpg')

# Generate pixel array from grayscale image 
pixels_gray = list(img_gray.getdata())

pixels_gray
```

### Normalize the single-digit pixel array from grayscale image: 

```python
# Open image and Convert to grayscale using pillow 
img_gray = Image.open('horsmil.jpg').convert('L')
img_gray.save('grayscale_horsmil.jpg')

# Generate pixel array from grayscale image 
pixels_gray = list(img_gray.getdata())

# Normalizing the pixel array using output from pillow: 

#Convert pixel list from pillow to an array 
pixels_gray = np.array(pixels_gray)

# confirm pixel range is 0-255
print('Before normalisation:')
print('Data Type: %s' % pixels_gray.dtype)
print('Min: %.3f, Max: %.3f' % (pixels_gray.min(), pixels_gray.max()))
print(pixels_gray, '\n')

# convert from integers to floats
pixels_gray = pixels_gray.astype('float32')

# normalize to the range 0-1
pixels_gray /= 255.0

# confirm the normalization
print('After normalisation:')
print('Data Type: %s' % pixels_gray.dtype)
print('Min: %.3f, Max: %.3f' % (pixels_gray.min(), pixels_gray.max()))
print(pixels_gray, '\n')
```
